```python
import pandas as pd
from fuzzywuzzy import fuzz
from tqdm import tqdm
```

```python
# Cargar datos
vin_cl = pd.read_csv('vin-cl (ford fusion).csv', sep=';')

# Ruta al archivo CSV
file_path = 'data-main (ford fusion) (1).csv'

# Nombres de columnas que proporcionas
column_names = ['col1', 'make', 'modelo', 'ano', 'engine', 'category', 'part', 'marca', 'col2', 'part_number']

# Cargar el archivo CSV con nombres de columnas personalizados
data_main = pd.read_csv(file_path, names=column_names, sep=';')

data_main = data_main.drop(['col1', 'category', 'part', 'col2'], axis=1)
```

```python
# Función para encontrar la mejor coincidencia en la segunda tabla
def encontrar_mejor_coincidencia(modelo, tabla):
    mejores_coincidencias = [(modelo2, fuzz.ratio(modelo, modelo2)) for modelo2 in tabla['modelo']]
    mejores_coincidencias.sort(key=lambda x: x[1], reverse=True)
    mejor_coincidencia = mejores_coincidencias[0]
    return mejor_coincidencia[0] if mejor_coincidencia[1] >= 80 else None
```

```python
# Fusionar las tablas en base a la mejor coincidencia
merged_data = pd.merge(data_main, vin_cl, how='left', left_on='Mejor_Coincidencia', right_on='modelo', suffixes=('_data_main', '_vin_cl'))
```
