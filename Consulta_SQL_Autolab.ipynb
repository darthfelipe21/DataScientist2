## Revisar el contenido de cada archivo

```python
# Almacenar en una variable todas las tablas
evaluacion= ['dbo-Brand.csv', 'dbo-Dealer.csv', 'dbo-Model.csv', 'dbo-Quotation.csv', 'dbo-QuotationProduct.csv']
```

```python
# Crear una función para recorrer el contenido de cada uno
def revisar_tablas(tablas):
    for i in tablas:
        with open(i, 'r') as file:
            for _ in range(10):
                print(file.readline())
            print('*'*20,'\n')

revisar_tablas(evaluacion)
```

## Realizar la consulta

```python
# Importar y conectar cada archivo
df_brand = pd.read_csv('dbo-Brand.csv', delimiter= ';')

df_model = pd.read_csv('dbo-Model.csv', delimiter= ';')

df_dealer = pd.read_csv('dbo-Dealer.csv', delimiter= ';')

df_quotation_product = pd.read_csv('dbo-QuotationProduct.csv', delimiter= ';')

df_quotation = pd.read_csv('dbo-Quotation.csv', delimiter=';')
```

```python
# Hacer los merge necesarios para optener una tabla unica
merged_df1 = pd.merge(df_quotation_product, df_quotation, on='QuotationId')

merged_df2 = pd.merge(merged_df1, df_dealer, on='DealerId')

merged_df3 = pd.merge(df_model, df_brand, on='BrandId')

final_df = pd.merge(merged_df2, merged_df3, on='ModelId')

final_df['ModelId'] = final_df['ModelId'].astype('int')

final_df
```

```python
# Crear DataFrame que contengan solo "Ecosport"
df= final_df[final_df['ModelId']== 156]
```

```python
df.isnull().sum()
```

```python
# Reorganizar y dejar columnas necesarias y que no esten llenas de nulos
df= df[['ModelId', 'DealerId', 'BrandId_x',
'ModelYearId', 'VersionId', 'QuotationCode', 'QuotationVinPlate',
'QuotationClientFullName', 'QuotationClientEmail', 'QuotationClientPhone',
'QuotationObservation', 'QuotationDate', 'GroupId', 'DealerName', 'DealerLanguage',
'DealerPhone', 'DealerLogo', 'DealerSubDomain', 'DealerSendCustomEmail', 'DealerToken',
'IsActive', 'DealerImage', 'DealerWhatsappIntegration', 'IsActiveSearchEngine',
'DealerHasApiIntegrationOrders', 'DealerHasApiIntegrationProducts', 'DealerIsIntegratedTransbank',
'DealerIsIntegratedGenet', 'DealerIsIntegratedMercadoPago', 'CommerceCodeTransbank',
'ApiKeyTransbank', 'DealerSearchType', 'DealerWhatsappIntegrationNotification',
'DealerWhatsappNumberIdentifierCodeNotification', 'DealerWhatsappPhoneNotification',
'ModelName','BrandName','QuotedProductId','QuotedProductName',]]
```

```python
# Guardar el DataFrame en una tabla de la base de datos
df.to_sql('ecosport', conn, index=False, if_exists='replace')

# Cerrar la conexión
conn.close()
```
